# 2) Unificar todos los dataset de 2023 y tener todas las columnas igual que el 2022

```python
# Unificar df_tabaquismo y df_nivel_actividad_fisica_2023

df_tabaquismo_nivel_actividad_fisica_2023 = pd.merge(df_tabaquismo, df_nivel_actividad_fisica_2023)
df_tabaquismo_nivel_actividad_fisica_2023
```

```python
df_tabaquismo_nivel_actividad_fisica_2023.isnull().sum()
```

```python
df_nivel_actividad_fisica_2022
```

# 3) Crear un dataset que tenga toda la data de 2022 y 2023

```python
df_tabaquismo_nivel_actividad_fisica_2022_y_2023 = pd.merge(df_tabaquismo_nivel_actividad_fisica_2023, df_nivel_actividad_fisica_2022, how='outer')
df_tabaquismo_nivel_actividad_fisica_2022_y_2023
```

___

# 4) Relacion entre consumo de alcohol y cancer (hint groupby mean)

```python
relacion_entre_consumo_alcohol_y_cancer = df_tabaquismo_nivel_actividad_fisica_2022_y_2023.groupby('Incidencia_Cancer').agg({'Consumo_Alcohol': 'mean'})
relacion_entre_consumo_alcohol_y_cancer
```

___

# 5) Relacion entre tabaquismo y cancer (hint groupby)

```python
relacion_entre_tabaquismo_y_cancer = df_tabaquismo_nivel_actividad_fisica_2022_y_2023.groupby('Incidencia_Cancer').agg({'Tabaquismo': 'mean'})
relacion_entre_tabaquismo_y_cancer
```

___

# 6) Correlacion entre tabaquismo y cancer

```python
df_tabaquismo_nivel_actividad_fisica_2022_y_2023['Tabaquismo'].corr(df_tabaquismo_nivel_actividad_fisica_2022_y_2023['Incidencia_Cancer'])
```

___

# 7) Correlacion entre consumo de alcohol y cancer

```python
df_tabaquismo_nivel_actividad_fisica_2022_y_2023['Consumo_Alcohol'].corr(df_tabaquismo_nivel_actividad_fisica_2022_y_2023['Incidencia_Cancer'])
```

# 8) ¿Cree que es el alcohol o es el tabaquismo el mayor causante de cancer?

#### Hay una correlacion mas fuerte entre tabaquismo y cancer, por lo tanto puedo decir que el tabaquismo es el mayor causante de cancer que el consumo de alcohol.
