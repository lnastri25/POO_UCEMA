```python
import pandas as pd
import numpy as n
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df1 = pd.DataFrame({'A': np.random.randint(1, 100, 10),
                    'B': np.random.randint(1, 100, 10),
                    'key': list(range(1, 6)) * 2})

df2 = pd.DataFrame({'C': np.random.randint(1, 100, 7),
                    'D': np.random.randint(1, 100, 7),
                    'key': list(range(2, 9))})

print("Sample data generated.")
```

```python
print("Inner Merge:")
inner_merged = pd.merge(df1,df2,how='inner')
print(inner_merged)
```

```python
print("Outer Merge:")
outer_merged = pd.merge(df1,df2,how='outer')
print(outer_merged)
```

```python
print("Left Merge:")
left_merged = pd.merge(df1,df2,how='left')
print(left_merged)
```

```python
print("Right Merge:")
right_merged = pd.merge(df1,df2,how='right')
print(right_merged)
```

```python
print("Correlation Matrix:")
corr_matrix = inner_merged.corr()
print(corr_matrix)
```

```python
print("Heatmap:")
sns.heatmap(corr_matrix, annot=True)
plt.show()
```
