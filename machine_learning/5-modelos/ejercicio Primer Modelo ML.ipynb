```python
# Crea la lista de características a continuación
feature_names = ["LotArea", "1stFlrSF", "2ndFlrSF", "FullBath", "BedroomAbvGr", "TotRmsAbvGrd"]

#  Seleccione los datos correspondientes a las funciones en feature_names
X = home_data[feature_names]
```

## Paso 3: especificar y ajustar el modelo
##### Cree un `DecisionTreeRegressor` y guárdelo iowa_model. Asegúrese de haber realizado la importación relevante desde sklearn para ejecutar este comando.

##### Luego ajuste el modelo que acaba de crear usando los datos en `X` e `y` que guardó arriba.

```python
from sklearn.tree import DecisionTreeRegressor
#especificar el modelo.
#Para la reproducibilidad del modelo, establezca un valor numérico para random_state al especificar el modelo
iowa_model = DecisionTreeRegressor(random_state=1)

# Fit the model
iowa_model.fit(X, y)
```

___

## Paso 4: hacer predicciones
##### Haga predicciones con el comando "predecir" del modelo utilizando "X" como datos. Guarde los resultados en una variable llamada "predicciones".

```python
predictions = iowa_model.predict(X.head())
print("Haciendo predicciones para las siguientes 5 casas:")
print(X.head())
print("Las Predicciones son:")
print(predictions)
```
