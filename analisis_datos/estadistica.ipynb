```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sns
from statistics import NormalDist
```

```python
male_df = pd.DataFrame([140, 145, 160, 190, 155, 165, 150, 190, 195, 138, 160, 155, 153, 145, 170, 175, 175, 170, 180, 135,
170, 157, 130, 185, 190, 155, 170, 155, 215, 150, 145, 155, 155, 150, 155, 150, 180, 160, 135, 160, 130, 155, 150, 148, 155,
150, 140, 180, 190, 145, 150, 164, 140, 142, 136, 123, 155],
 columns=['weight'])
```

```python
male_df['sex'] = 'male'
female_df = pd.DataFrame([140, 120, 130, 138, 121, 116, 125, 145, 150, 112, 125, 130, 120, 130, 131, 120, 118, 125, 135, 125,
118, 122, 115, 102, 115, 150, 110, 116, 108, 95, 125, 133, 110, 150, 108],
 columns=['weight'])
female_df['sex'] = 'female'
```

```python
weights_df = pd.concat([male_df, female_df], ignore_index=True) # me arma uno entero con 2 dfâ€™s concatenados siempre y cuando los valores sean los mismos
weights_df.sample(5)
```

___

# Histograma

```python
ax = sns.histplot(weights_df["weight"], bins=len(weights_df))
ax.set_xlabel("Weight (in lbs)")
plt.show()

# bins: cantidad de barras
# ax: eje x
```

### â†’ Un histograma es una representaciÃ³n precisa de la distribuciÃ³n de datos numÃ©ricos. 
### â†’ Es una estimaciÃ³n de la distribuciÃ³n de probabilidad de una variable continua.


___

# Histogram Bins

```python
f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 4))
ax1.hist(weights_df["weight"], bins=[95, 105, 115, 125, 135, 145, 155, 165, 175, 185, 195, 205, 215])
ax2.hist(weights_df["weight"], bins=12)
sns.histplot(weights_df["weight"], bins=12, ax=ax3, kde=True)
plt.show()
```

___

# Cumulative Plots

```python
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 4))
ax1.hist(weights_df["weight"], bins=25, cumulative=True, density=True)
sns.histplot(weights_df["weight"], binwidth=5, ax=ax2, kde=True,
 cumulative=True, stat="percent");
```

### â†’ Alternativamente, podemos trazar el recuento de pesos inferiores a un determinado valor. 
### â†’ En lugar de los recuentos, tambiÃ©n podemos trazar la densidad (sumas hasta 1) o el porcentaje (sumas hasta 100).
