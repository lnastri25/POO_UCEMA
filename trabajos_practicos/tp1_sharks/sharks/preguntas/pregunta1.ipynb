# 1) ¿Los tiburones matan más a gente mayor?

```python
tiburones1['Fatal (Y/N)'] = tiburones1['Fatal (Y/N)'].replace('F', 'Y')
```

```python
fatal = tiburones1['Fatal (Y/N)'] == 'Y'
```

```python
df_1 = tiburones1[fatal]
df_1
```

```python
# df_2['Age'].value_counts()
```

```python
df_2 = df_1['Age']
df_2
```

```python
valores_numericos = pd.to_numeric(df_2, errors='coerce')
```

```python
valores_numericos = valores_numericos.dropna()
valores_numericos
```

```python
df_mayores_de_60 = valores_numericos[valores_numericos >= 60].count()
df_mayores_de_60
```

```python
df_menores_de_60 = valores_numericos[valores_numericos < 60].count()
df_menores_de_60
```

___

    ➢ Conclusión:

    Murieron más personas que son menores de 60 años que aquellas que su edad supera los 60 años.
