```python
import pandas as pd
import matplotlib.pyplot as plt

# Filtrar las filas con valores "Sí" o "No" en la columna "El cliente se encontró satisfecho con el resultado?"
df_filtrado = opiniones[opiniones['El cliente se encontró satisfecho con el resultado?'].isin(['Sí', 'No'])]

# Convertir la columna "Edad" a tipo numérico (si aún no lo está)
df_filtrado['Edad'] = pd.to_numeric(df_filtrado['Edad'], errors='coerce')

# Eliminar las filas con valores no válidos en la columna "Edad"
df_filtrado = df_filtrado.dropna(subset=['Edad'])

# Limitar el rango de edades entre 20 y 80 años
df_filtrado = df_filtrado[(df_filtrado['Edad'] >= 20) & (df_filtrado['Edad'] <= 80)]

# Calcular estadísticas de edad
edad_promedio = df_filtrado['Edad'].mean()
mediana = df_filtrado['Edad'].median()

# Crear un histograma de edades para clientes satisfechos/insatisfechos
plt.figure(figsize=(10, 6))
plt.hist(df_filtrado['Edad'], bins=10, range=(20, 80), color='skyblue', edgecolor='black')
plt.title('Distibución de clientes entre 20 y 80 años')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')

# Agregar leyenda con mediana y promedio sin el color
legend_label = f'Promedio: {edad_promedio:.2f}\nMediana: {mediana:.2f}'
plt.legend([legend_label], loc='upper right', handlelength=0)

plt.show()
```
