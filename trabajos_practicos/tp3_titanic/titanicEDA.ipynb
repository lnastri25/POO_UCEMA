# Titanic Disaster

Vamos a explorar el dataset de titanic


Acerca de este archivo
El hundimiento del Titanic es uno de los naufragios más infames de la historia.

El 15 de abril de 1912, durante su viaje inaugural, el RMS Titanic, ampliamente considerado “insumergible”, se hundió después de chocar con un iceberg. Desafortunadamente, no había suficientes botes salvavidas para todos a bordo, lo que provocó la muerte de 1.502 de los 2.224 pasajeros y tripulantes.

Si bien hubo algún elemento de suerte involucrado en la supervivencia, parece que algunos grupos de personas tenían más probabilidades de sobrevivir que otros.

En este desafío, le pedimos que construya un modelo predictivo que responda a la pregunta: "¿qué tipo de personas tenían más probabilidades de sobrevivir?" utilizando datos de los pasajeros (es decir, nombre, edad, sexo, clase socioeconómica, etc.).

Columnas

*   PassengerId -> Passenger ID
*   Survived -> Sobreviviente 0 = No, 1 = Sí
*   Pclass -> Ticket clase: 1 = 1st, 2 = 2nd, 3 = 3rd
*   Name -> Nombre Pasajero
*   Sex -> Genero
*   Age -> Edad
*   SibSp -> Número de hermanos / cónyuges a bordo del Titanic
*   Parch -> Número de padres/hijos a bordo del Titanic
*   Ticket -> Número Ticket
*   Fare -> Tarifa de pasajero
*   Cabin -> # de Cabina
*   Embarked -> Embarcado implica desde dónde montó el viajero. Hay tres valores posibles para Embark: Southampton, Cherbourg y Queenstown.









```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
titanic_df = pd.read_csv('https://wagon-public-datasets.s3.amazonaws.com/Machine%20Learning%20Datasets/ML_Titanic_dataset.csv')
```

```python
titanic_df
```

# 1) De los sobrevivientes:

```python
titanic_sobrevivientes = titanic_df[titanic_df['Survived'] == 1].count()['Survived']
print("La cantidad de sobrevivientes del Titanic es de", titanic_sobrevivientes, "personas.")
```

```python
sobrevivientes = titanic_df[titanic_df['Survived'] == 1]
sobrevivientes
```

# Clases en las que viajaron

```python
titanic_clases_sobrevivientes = sobrevivientes["Pclass"].value_counts()
print("Sobrevivieron", titanic_clases_sobrevivientes[1], "de primera clase,", titanic_clases_sobrevivientes[2], "de segunda clase y", titanic_clases_sobrevivientes[3], "de tercera clase.")
```

# Ahora con las Edades

```python
titanic_edades_sobrevivientes = sobrevivientes["Age"].value_counts()
titanic_edades_sobrevivientes.head(10)
```

```python
edades_primera_clase_sobrevivientes = sobrevivientes[sobrevivientes["Pclass"] == 1]["Age"].median()
print("La mediana de los que sobrevivieron en primera clase es de", int(edades_primera_clase_sobrevivientes), "años.")
```

```python
edades_segunda_clase_sobrevivientes = sobrevivientes[sobrevivientes["Pclass"] == 2]["Age"].median() 
print("La mediana de los que sobrevivieron en segunda clase es de", int(edades_segunda_clase_sobrevivientes), "años.")
```

```python
edades_tercera_clase_sobrevivientes = sobrevivientes[sobrevivientes["Pclass"] == 3]["Age"].median()
print("La mediana de los que sobrevivieron en tercera clase es de", int(edades_tercera_clase_sobrevivientes), "años.")
```

# Ahora con los Géneros

```python
titanic_generos_sobrevivientes = sobrevivientes["Sex"].value_counts()
print("Podemos suponer que porque sobrevivieron", titanic_generos_sobrevivientes[1], "hombres y sobrevivieron", titanic_generos_sobrevivientes[0], "mujeres, las mujeres tenían más probabilidades de sobrevivir, o quizás les dieron mayor ""prioridad"" a ellas para subir a los botes de rescate antes que a los hombres.")
```

# Lugares de Embarque

```python
titanic_embarcados_sobrevivientes = sobrevivientes["Embarked"].value_counts()
print("Sobrevivieron", titanic_embarcados_sobrevivientes[0], "de los que embarcaron en Southampton,", titanic_embarcados_sobrevivientes[1], "de los que embarcaron en Cherbourg y", titanic_embarcados_sobrevivientes[2], "de los que embarcaron en Queenstown.")
```

```python
cabina_sobrevivientes = sobrevivientes["Fare"].value_counts(ascending=False)
cabina_sobrevivientes
```

# Hermanos/cónyuges y Padres/hijos y Tutores

```python
hermanos_sobrevivientes = sobrevivientes["SibSp"].value_counts()
hermanos_sobrevivientes
```

```python
familias_sobrevivientes = sobrevivientes["Parch"].value_counts()
familias_sobrevivientes
```
