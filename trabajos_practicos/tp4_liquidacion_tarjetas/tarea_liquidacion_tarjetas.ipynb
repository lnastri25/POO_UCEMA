### 6- ¿Cuanto dinero se cobro por mes? Cree un lineplot para gráficarlo


```python
# paso la fecha_de_pago a mes

tarjetas["mes"] = tarjetas["fecha_de_pago"].dt.month
tarjetas
```

```python
cobro_mensual = tarjetas.groupby('mes').agg(
    cobros_mensuales = ('importe_concepto', 'sum')
).reset_index()
cobro_mensual
```

```python
meses = cobro_mensual['mes']
cobros = cobro_mensual['cobros_mensuales']
plt.plot(meses, cobros, color='red', marker='o')
plt.title('Dinero cobrado por mes')
plt.xlabel('Mes')
plt.ylabel('Dinero cobrado')
plt.show()
```

### 11- Quedarse solo con tarjeta visa. Crear una columna que sume 'VENTAS C/DESCUENTO CONTADO' con 'VENTAS C/DTO CUOTAS FINANC. OTORG.'

```python
df1 = tarjetas.copy()
condicion_visa = df1['cod_marca_producto'].str.contains('VISA')
tarjetas_visa = df1[condicion_visa]
visa_por_desc_creaing = tarjetas_visa.pivot_table(index=["fecha_de_pago","cod_comercio"],columns="desc_creaing",values="importe_concepto")
visa_por_desc_creaing
```

```python
visa_por_desc_creaing[['VENTAS C/DESCUENTO CONTADO                  ','VENTAS C/DTO CUOTAS FINANC. OTORG.          ']]
```

```python
visa_por_desc_creaing['ventas_con_descuento'] = visa_por_desc_creaing['VENTAS C/DESCUENTO CONTADO                  '] + visa_por_desc_creaing['VENTAS C/DTO CUOTAS FINANC. OTORG.          ']
pd.DataFrame(visa_por_desc_creaing['ventas_con_descuento']).reset_index()
```
